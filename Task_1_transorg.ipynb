```python
# To access the 7000+ records from imdb 
pages=np.arange(1,7206,50)
```

```python
# Code to access data from imdb like movie_name,released year and reviews related to that movie

# This will display data in form of a dataframe
result_df = pd.DataFrame()


# Applying the loop to to access 50 records each time and then navigating the next page.
# We can even access 100000+ records or more than it for that we have to change only the base_url for it the rest of the code will remain same.
# In this particular sceanrio I have taken up only 7206 titles.
for base_url in pages:
    base_url = requests.get("https://www.imdb.com/search/title/?countries=us&title_type=feature&num_votes=10000,&sort=user_rating,desc&count=50&start="+str(base_url)+"&ref_=adv_nxt")
    url_header_for_reviews = 'https://www.imdb.com'
    url_tail_for_reviews = 'reviews?ref_=tt_urv'
    html_soup = BeautifulSoup(base_url.text, 'html.parser')
    movie_containers = html_soup.find_all('div', attrs = {'class': 'lister-item mode-advanced'})
    sleep(randint(2,8))
# Extract data from individual movie container
    for container in movie_containers:
# Reviews extracting
        num_reviews = 1
        # Getting last piece of link puzzle for a movie reviews` link
        url_middle_for_reviews = container.find('a')['href']
        # Opening reviews page of a concrete movie
        response_reviews = get(url_header_for_reviews + url_middle_for_reviews + url_tail_for_reviews)
        reviews_soup = BeautifulSoup(response_reviews.text, 'html.parser')
        # Searching all reviews
        reviews_containers = reviews_soup.find_all('div', class_ = 'imdb-user-review')
        # Check if actual number of reviews is less than target one
        if len(reviews_containers) < num_reviews:
            num_reviews = len(reviews_containers)
        # Looping through each review and extracting title and body
        reviews_titles = []
        reviews_bodies = []
        for review_index in range(num_reviews):
            review_container = reviews_containers[review_index]
            review_title = review_container.find('a', class_ = 'title').text.strip()
            review_body = review_container.find('div', class_ = 'text').text.strip()
            reviews_titles.append(review_title)
            reviews_bodies.append(review_body)
# Movie name
        name = container.h3.a.text
        names = [name for i in range(num_reviews)]
# Movie year
        year = container.h3.find('span', class_ = 'lister-item-year').text.replace('(','').replace(')','')
        years = [year for i in range(num_reviews)]

# Gathering up scraped data into result_df
        if result_df.empty:
            result_df = pd.DataFrame({'movie': names,'year': years,'review_title': reviews_titles,'review_body': reviews_bodies})
        elif num_reviews > 0:
            result_df = result_df.append(pd.DataFrame({'movie': names,'year': years,'review_title': reviews_titles,'review_body': reviews_bodies}))
```
